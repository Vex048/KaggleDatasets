```python
import pandas as pd
```

```python
df_trainset = pd.read_csv("datasets/train.csv")
df_testset = pd.read_csv("datasets/test.csv")
```

```python
df_trainset.shape
```

```python
# Data preproccesing 
```

```python
df_trainset.isnull().sum()
```

```python
print(df_trainset.HomePlanet.value_counts())
print(df_trainset.CryoSleep.value_counts())
print(df_trainset.Destination.value_counts())
print(df_trainset.Cabin.value_counts())
```

```python
df_trainset = df_trainset.dropna()
```

```python
Y_train = df_trainset['Transported']
Y_train
```

```python
df_trainset = df_trainset.drop('Name',axis=1)
df_trainset = df_trainset.drop('Transported',axis=1)
```

```python
df_trainset
```

```python
categorical_columns = ['HomePlanet','CryoSleep','Destination']
df_trainset = pd.get_dummies(df_trainset,columns=categorical_columns)
df_trainset
```

```python
df_trainset = df_trainset.replace({True: 1,False: 0})
```

```python
df_trainset
```

```python
columns_to_normalize = ['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = df_trainset
X_train[['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']] = scaler.fit_transform(X_train[['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']])
X_train
```

```python
X_train.Cabin.value_counts().head(10)
```

```python
Cabin_values = X_train.Cabin.value_counts()
len(X_train.Cabin.unique())
```

```python
Cabin_values
```

```python
import numpy as np

X_train['Cabin'] = np.where(Cabin_values[X_train['Cabin']] < 2,'other',X_train['Cabin'])
```

```python
X_train
```

```python
X_train
```

```python
X_train = pd.get_dummies(X_train,columns=['Cabin'])
X_train = X_train.replace({True: 1,False: 0})
```

```python
X_train
```

```python
X_train.drop('PassengerId',axis=1)
```

```python
X_train.isnull().sum().sum()
```

```python
X_train
```

```python
print(X_train.shape)
print(df_testset.shape)
print(Y_train.shape)
print(df_testset)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_train, Y_train, test_size=0.2, random_state=42)
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=10000)
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_train)
print(y_train.shape)
print(y_predict.shape)
from sklearn.metrics import confusion_matrix
matrix=confusion_matrix(y_train,y_predict)
print(matrix)
accuracy = (matrix[0][0] + matrix [1][1]) / (matrix.sum())
print(accuracy)
```

```python
# Confusion matrix for train set and cabin < 1 is - 
#[[2034  592]
 #[ 531 2127]]
#Accuracy is 0.7874716124148372 == 79%
```

```python
y_predict_test=model.predict(X_test)
print(y_test.shape)
print(y_predict_test.shape)
matrix_test=confusion_matrix(y_test,y_predict_test)
print(matrix_test)
accuracy = (matrix_test[0][0] + matrix_test [1][1]) / (matrix_test.sum())
print(accuracy)
```

```python
# Confusion matrix for test set cabin < 1  is - 
#[[496 157]
#[132 537]]
#Accuracy is 0.7874716124148372 == 78%
```

```python
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam
model_keras = keras.Sequential([
    layers.Dense(units = 5,activation='relu'),
    layers.Dense(units = 5,activation='relu'),
    layers.Dense(units = 1,activation='sigmoid')
])
model_keras.compile(loss = BinaryCrossentropy(),optimizer=Adam(learning_rate=0.01),metrics=['accuracy'])
model_keras.fit(X_train,y_train,epochs = 100)
```

```python
evaluation_first_model_train = model_keras.evaluate(X_train,y_train)
```

```python
#Accuracy 0.8638 - loss: 0.3011 on train set 
```
